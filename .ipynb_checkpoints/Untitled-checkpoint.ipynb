```python
import os
from glob import glob
import matplotlib.pyplot as plt
import rasterio as rio
from rasterio.plot import plotting_extent
import geopandas as gpd
import earthpy as et
import earthpy.spatial as es
import earthpy.plot as ep
```

```python
# Get sample data from EarthPy and set working directory
data_path = et.data.get_data("vignette-landsat")
os.chdir(os.path.join(et.io.HOME, "earth-analytics"))

# Get list of bands and sort by ascending band number
landsat_bands_data_path = "data/vignette-landsat/LC08_L1TP_034032_20160621_20170221_01_T1_sr_band*[1-7]*.tif"
stack_band_paths = glob(landsat_bands_data_path)
stack_band_paths.sort()

# Create image stack and apply nodata value for Landsat
arr_st, meta = es.stack(stack_band_paths, nodata=-9999)
```

```python
fig, ax = plt.subplots(figsize=(19.80,19.80))

# Plot red, green, and blue bands, respectively
ep.plot_rgb(arr_st, rgb=(3, 2, 1), ax=ax, title="Landsat 8 RGB Image")
plt.savefig('Landsat8rgb.png',bbox_inches='tight', dpi=300)
# plt.show()
```

```python
fig, ax = plt.subplots(figsize=(19.80,19.80))
#Plot after linear stretching
ep.plot_rgb(arr_st, rgb=(3, 2, 1), ax=ax, stretch=True,str_clip=1,title="Landsat 8 RGB Image with Stretch applied")
plt.savefig('landsat8-LinearStretched.png',bbox_inches='tight',dpi=300)
# plt.show()
```

```python
fig, ax = plt.subplots(figsize=(19.80, 19.80))

# Plot red, green, and blue bands, respectively
ep.plot_rgb(arr_st, rgb=(4, 3, 2), ax=ax, title="Landsat 8 Infrared Image")
plt.savefig('Landsat8infra.png',bbox_inches='tight',dpi=300)
# plt.show()
```

```python
fig, ax = plt.subplots(figsize=(19.80, 19.80))

# Plot red, green, and blue bands, respectively
ep.plot_rgb(arr_st, rgb=(4, 3, 2), ax=ax, stretch=True,str_clip=1,title="Landsat 8 Infrared Image Stretched")
plt.savefig('Landsat8infra-Linear Stretched.png',bbox_inches='tight', dpi=300)
# plt.show()
```
